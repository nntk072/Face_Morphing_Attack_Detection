```python
def get_video_files(input_directory):
    """Recursively collects video files from the given directory and its sub-directories."""
    video_files = []
    for root, dirs, files in os.walk(input_directory):
        for file in files:
            if file.endswith(('.mp4', '.avi', '.mov', '.mkv')):
                video_files.append(os.path.join(root, file))
    return video_files

def extract_central_frame(video_path, output_directory):
    """Extracts the central frame from the given video and saves it to the output directory."""
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error opening video file {video_path}")
        return

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    central_frame_index = total_frames // 2

    cap.set(cv2.CAP_PROP_POS_FRAMES, central_frame_index)
    ret, frame = cap.read()
    if ret:
        video_name = os.path.basename(video_path)
        frame_name = os.path.splitext(video_name)[0] + '_central_frame.jpg'
        frame_path = os.path.join(output_directory, frame_name)
        cv2.imwrite(frame_path, frame)
        print(f"Saved central frame of {video_name} to {frame_path}")
    else:
        print(f"Failed to read frame from {video_path}")

    cap.release()

def process_videos(input_directory, output_directory):
    """Processes all video files in the input directory to extract and save central frames."""
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    video_files = get_video_files(input_directory)
    for video_file in video_files:
        extract_central_frame(video_file, output_directory)

if __name__ == "__main__":
    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/attack"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/attack1"
    # process_videos(input_directory, output_directory)

    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/real"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/real1"
    # process_videos(input_directory, output_directory)

    # # Test directory
    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/attack"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/attack1"
    # process_videos(input_directory, output_directory)
    
    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/real"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/real1"
    # process_videos(input_directory, output_directory)
    
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/attack"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/attack1"
    process_videos(input_directory, output_directory)
    
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/real"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/real1"
    process_videos(input_directory, output_directory)
    
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/attack"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/attack1"
    process_videos(input_directory, output_directory)
    
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/real"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/real1"
    process_videos(input_directory, output_directory)
```

```python
# Function to detect and crop faces in an image using dlib's face detector
def detect_and_crop_faces(image, output_folder, file_base_name, scale_factor=1.7, target_size=(224, 224)):
    # Initialize dlib's face detector
    detector = dlib.get_frontal_face_detector()
    
    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the image using dlib
    faces = detector(gray)
    
    # If no faces are detected with dlib, use Viola-Jones method
    if len(faces) == 0:
        print(f"No faces detected with dlib, trying Viola-Jones method.")
        # Load OpenCV's pre-trained Haar Cascade classifier for face detection
        # haar_cascade_path = '/u/45/muhammu2/unix/four datasets/CASIA Dataset/training set/haarcascade_frontalface_default.xml'  # Update with actual path
        haar_cascade_path = 'c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/haarcascade_frontalface_default.xml'  # Update with actual path
        face_cascade = cv2.CascadeClassifier(haar_cascade_path)
        
        # Detect faces using Haar Cascades with increased bounding box size
        faces_cv = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(int(30*scale_factor), int(30*scale_factor)))
        faces = [dlib.rectangle(x, y, x + w, y + h) for (x, y, w, h) in faces_cv]
    
    # If still no faces are detected, skip this image
    if len(faces) == 0:
        print(f"No faces detected with both methods.")
        return
    
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Iterate over detected faces and crop them
    for i, face in enumerate(faces):
        # Get bounding box coordinates
        (x, y, w, h) = (face.left(), face.top(), face.width(), face.height())
        
        # Enlarge bounding box
        new_x = max(0, x - int((scale_factor - 1) * w / 2))
        new_y = max(0, y - int((scale_factor - 1) * h / 2))
        new_w = int(w * scale_factor)
        new_h = int(h * scale_factor)
        
        # Crop the enlarged face region
        cropped_face = image[new_y:new_y+new_h, new_x:new_x+new_w]
        
        # Check if the cropped face is not empty
        if cropped_face is not None and cropped_face.size > 0:
            # Resize the cropped face to target size (224x224)
            resized_face = cv2.resize(cropped_face, target_size, interpolation=cv2.INTER_CUBIC)
            
            # Save the resized face
            face_output_path = os.path.join(output_folder, f"{file_base_name}_face_{i}.jpg")
            cv2.imwrite(face_output_path, resized_face)
            print(f"Face {i} saved as: {face_output_path}")
        else:
            print(f"Error: Cropped face is empty, face {i}")

# Function to extract the central frame from a video and detect and crop faces
def extract_central_frame(video_path, output_directory):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error opening video file {video_path}")
        return

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    central_frame_index = total_frames // 2

    cap.set(cv2.CAP_PROP_POS_FRAMES, central_frame_index)
    ret, frame = cap.read()
    if ret:
        video_name = os.path.basename(video_path)
        file_base_name = os.path.splitext(video_name)[0]
        
        # Detect and crop faces from the central frame
        detect_and_crop_faces(frame, output_directory, file_base_name)
    else:
        print(f"Failed to read frame from {video_path}")

    cap.release()


if __name__ == "__main__":
    # # Train directory
    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/attack"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/crop_attack"
    # process_videos(input_directory, output_directory)
    
    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/real"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/train/crop_real"
    # process_videos(input_directory, output_directory)
    
    # # Test directory
    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/attack"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/crop_attack"
    # process_videos(input_directory, output_directory)
    
    # input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/real"
    # output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Replay Attack dataset/test/crop_real"
    # process_videos(input_directory, output_directory)

    # Train directory
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/attack"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/crop_attack"
    process_videos(input_directory, output_directory)
    
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/real"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/train/crop_real"
    process_videos(input_directory, output_directory)
    
    # Test directory
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/attack"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/crop_attack"
    process_videos(input_directory, output_directory)
    
    input_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/real"
    output_directory = "c:/Users/nguyenl37/OneDrive - Aalto University/University/First year/Project/Face_Morphing_Attack_Detection/Casia_dataset/test/crop_real"
    process_videos(input_directory, output_directory)
```
